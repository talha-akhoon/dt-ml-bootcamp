```python
#q7
X = csv[csv.origin == 'Asia'][['vehicle_weight','model_year']].head(7)
X_T = X.T
X_T_X = X_T.dot(X)

import numpy as np
X_T_X_inv = np.linalg.inv(X_T_X.values)
X_T_X_inv

y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])


# Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
w = X_T_X_inv.dot(X_T).dot(y)
w.sum()
```
