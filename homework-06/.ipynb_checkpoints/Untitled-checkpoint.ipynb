```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split

#q1
pd.__version__

#q2
df = pd.read_csv("https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv")
```

```python
target = 'fuel_efficiency_mpg'
y = df[target]
X = df.drop(columns=[target])

X.head()
```

```python
X.isnull().describe()
```

```python
X['num_cylinders'] = X['num_cylinders'].fillna(0)
```

```python
X['horsepower'] = X['horsepower'].fillna(0)
```

```python
X['acceleration'] = X['acceleration'].fillna(0)
X['num_doors'] = X['num_doors'].fillna(0)
```

```python
dv = DictVectorizer(sparse=True)
df_train, df_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
df_train, df_val, y_train, y_val = train_test_split(df_train, y_train, test_size=0.25, random_state=1)
```

```python
X_train = dv.fit_transform(df_train.to_dict(orient='records'))

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth=1)
dtr.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text

print(export_text(dtr, feature_names=dv.get_feature_names_out()))

#Q1 answer: vehicle_weight
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
rfr.fit(X_train, y_train)

X_val = dv.transform(df_val.to_dict(orient='records'))

y_pred = rfr.predict(X_val)

from sklearn.metrics import mean_squared_error
import numpy as np

np.sqrt(mean_squared_error(y_val, y_pred))

#Q2 answer: 0.45
```

```python
start = 10
finish = 210
step = 10

rmse_n = []
for n in range(start, finish, step):
    rfr = RandomForestRegressor(n_estimators=n, random_state=1, n_jobs=-1)
    rfr.fit(X_train, y_train)
    X_val = dv.transform(df_val.to_dict(orient='records'))
    y_pred = rfr.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))

    rmse_n.append([n, rmse])

df_rmse_n = pd.DataFrame(rmse_n, columns=['n', 'rmse'])

import matplotlib.pyplot as plt

%matplotlib inline
plt.plot(df_rmse_n['n'], df_rmse_n['rmse'])
plt.xlabel('n_estimators')
plt.ylabel('rmse')

plt.show()

df_rmse_n

#Q3 answer: 200 (0.443) 3 d.p, 80 (0.445) 3 d.p
```

```python
start = 10
finish = 210
step = 10
mean_rmse_d = []
for d in [10, 15, 20, 25]:
    rmse_list = []
    for n in range(start, finish, step):
        rfr = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=1, n_jobs=-1)
        rfr.fit(X_train, y_train)
        X_val = dv.transform(df_val.to_dict(orient='records'))
        y_pred = rfr.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
        rmse_list.append(rmse)
    mean_rmse_d.append([d, np.mean(rmse_list)])

df_rmse_d = pd.DataFrame(mean_rmse_d, columns=['d', 'mean_rmse'])

df_rmse_d
```

```python

plt.plot(df_rmse_d['d'], df_rmse_d['mean_rmse'])
plt.xlabel('d (Max depth)')
plt.ylabel('Mean RMSE')
plt.title('Mean RMSE vs Max Depth')
plt.show()

#Q4 answer: 10
```

```python
binary_feature_importannce = rfr.feature_importances_
feature_names = dv.get_feature_names_out()
together = zip(binary_feature_importannce, feature_names)
pd.DataFrame(together, columns=['binary_feature_importannce', 'feature_names'])
#Q5 answer: vehicle_weight
```

```python
import xgboost as xgb


xgb_params = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'reg:squarederror',
    'nthread': 8,
    'seed': 1,
    'verbosity': 1
}

features = list(feature_names)
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)

result = {}
for eta in [0.3, 0.1]:
    xgb_params['eta'] = eta
    model = xgb.train(xgb_params, dtrain, num_boost_round=100)
    y_pred = model.predict(dval)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    result[eta] = rmse


result

#Q6 answer: 0.1
```
